## countplot

```python
  plt.figure(figsize=(8,6))
  purchased = data["Purchased"]
  total = len(purchased)*1.
  ax=sns.countplot(x="Purchased", data=data)
  for p in ax.patches:
    ax.annotate('{:.1f}%'.format(100*p.get_height()/total), (p.get_x()+0.1, p.get_height()+5))

  #put 11 ticks (therefore 10 steps), from 0 to the total number of rows in the dataframe
  ax.yaxis.set_ticks(np.linspace(0, total, 11))
  #adjust the ticklabel to the desired format, without changing the position of the ticks.
  ax.set_yticklabels(map('{:.1f}%'.format, 100*ax.yaxis.get_majorticklocs()/total))
  ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
  # ax.legend(labels=["no","yes"])
  plt.show()
```

```python
def countplot_withY(label, data):
    plt.figure(figsize=(20,10))
    Y = data[label]
    total = len(Y)*1.
    ax=sns.countplot(x=label, data=data, hue="Purchased")
    for p in ax.patches:
        ax.annotate('{:.1f}%'.format(100*p.get_height()/total), (p.get_x()+0.1, p.get_height()+5))

    #put 11 ticks (therefore 10 steps), from 0 to the total number of rows in the dataframe
    ax.yaxis.set_ticks(np.linspace(0, total, 11))
    #adjust the ticklabel to the desired format, without changing the position of the ticks.
    ax.set_yticklabels(map('{:.1f}%'.format, 100*ax.yaxis.get_majorticklocs()/total))
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
    # ax.legend(labels=["no","yes"])
    plt.show()
```

```python
countplot_withY('Country',data)
```

## boxplot

```python
def distplot(label,data):
    sns.boxplot(data=data, x='Purchased', y=label)
    plt.show()
    plt.figure(figsize=(5,4))
    sns.distplot(data[label])
    plt.show()
```

```python
distplot('Age',data)
```

## OneHotEncoder

```python
data_ready = data.copy()
```

```python
data_ready
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# create OneHotEncoder object
encoder = OneHotEncoder()

cat_cols = ['Country']

# fit and transform color column
one_hot_array = encoder.fit_transform(data_ready[['Country']]).toarray()

try:
  encoded_feature_names = encoder.get_feature_names()
except AttributeError:
  # Fallback for older scikit-learn versions (manual feature name creation)
  categories = encoder.categories_[0]
  encoded_feature_names = [f"Country_{category}" for category in categories]

# create new dataframe from numpy array
one_hot_encoder = pd.DataFrame(one_hot_array, columns=encoded_feature_names)

data_ready = data_ready.drop(cat_cols ,axis=1)
data_ready = pd.concat([one_hot_encoder,data_ready], axis=1)
# Encode target value
data_ready['Purchased'] = data_ready['Purchased'].apply(lambda x: 1 if x == 'Yes' else 0)


```

```python
data_ready
```

```python
path = 'DataSet/ModifiedDataReady.csv'
data_ready.to_csv(path, index=False)
```

## Split the data

```python
# Select Features
feature = data_ready.drop('Purchased', axis=1)

# Select Target
target = data_ready['Purchased']

# Set Training and Testing Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature , target, 
                                                    shuffle = True, 
                                                    test_size=0.2, 
                                                    random_state=1)

# Show the Training and Testing Data
print('Shape of training feature:', X_train.shape)
print('Shape of testing feature:', X_test.shape)
print('Shape of training label:', y_train.shape)
print('Shape of training label:', y_test.shape)
```

## StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_cols = ['Age','Salary']
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])

print(X_train.shape)
X_train.head(10)
```

## PCA

```python
from sklearn.decomposition import PCA  # 1. Choose the model class
model = PCA(n_components=2)            # 2. Instantiate the model with hyperparameters
model.fit(feature)                      # 3. Fit to data. Notice y is not specified!
X_2D = model.transform(feature)         # 4. Transform the data to two dimensions
```

```python
data_ready['PCA1'] = X_2D[:, 0]
data_ready['PCA2'] = X_2D[:, 1]
```

```python
import seaborn as sns

sns.lmplot(x="PCA1", y="PCA2", hue="Purchased", data=data_ready, fit_reg=False)
```
