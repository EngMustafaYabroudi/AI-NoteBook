```python
import pandas as pd
import numpy as np
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

## Load dataset

```python
data_ready = pd.read_csv('DataSet/ModifiedDataReady.csv')
```

```python
data_ready.head(10)
```

## Select Features

```python
# Select Features
feature = data_ready.drop('Purchased', axis=1)

# Select Target
target = data_ready['Purchased']

# Set Training and Testing Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature , target, 
                                                    shuffle = True, 
                                                    test_size=0.2, 
                                                    random_state=1)

# Show the Training and Testing Data
print('Shape of training feature:', X_train.shape)
print('Shape of testing feature:', X_test.shape)
print('Shape of training label:', y_train.shape)
print('Shape of training label:', y_test.shape)
```

```python
from sklearn.decomposition import PCA  # 1. Choose the model class
model = PCA(n_components=2)            # 2. Instantiate the model with hyperparameters
model.fit(feature)                      # 3. Fit to data. Notice y is not specified!
X_2D = model.transform(feature) 
```

```python
data_ready['PCA1'] = X_2D[:, 0]
data_ready['PCA2'] = X_2D[:, 1]
```

```python
data_ready.head(10)
```

```python
import seaborn as sns

sns.lmplot(x="PCA1", y="PCA2", hue="Purchased", data=data_ready, fit_reg=False)
```

## evaluate_model 

```python
def evaluate_model(model, x_test, y_test):
    from sklearn import metrics

    # Predict Test Data 
    y_pred = model.predict(x_test)

    # Calculate accuracy, precision, recall, f1-score, and kappa score
    acc = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    f1 = metrics.f1_score(y_test, y_pred)
    kappa = metrics.cohen_kappa_score(y_test, y_pred)

    # Calculate area under curve (AUC)
    y_pred_proba = model.predict_proba(x_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)

    # Display confussion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)

    return {'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1, 'kappa': kappa, 
            'fpr': fpr, 'tpr': tpr, 'auc': auc, 'cm': cm}
```

```python
from sklearn.cluster import KMeans     # 1. Choose the model class
model = KMeans(n_clusters=2)          # 2. Instantiate the model with hyperparameters
model.fit(feature)                    # 3. Fit to data. Notice y is not specified!
y_kmeans = model.predict(feature)        # 4. Determine cluster labels
```

```python
data_ready['cluster'] = y_kmeans
```

```python
sns.lmplot(x="PCA1", y="PCA2",data=data_ready, hue='Purchased',
           col='cluster', fit_reg=False)
```

```python
from sklearn.dummy import DummyClassifier

# Negative class (0) is most frequent
dummy_majority = DummyClassifier(strategy = 'most_frequent')  #stratified , uniform
dummy_majority.fit(X_train, y_train)
```

```python
dm_eval = evaluate_model(dummy_majority, X_test, y_test)

# Print result
print('Accuracy:', dm_eval['acc'])
print('Precision:', dm_eval['prec'])
print('Recall:', dm_eval['rec'])
print('F1 Score:', dm_eval['f1'])
print('Area Under Curve:', dm_eval['auc'])
print('Confusion Matrix:\n', dm_eval['cm'])
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=0)
lr.fit(X_train, y_train)
```

```python
# Evaluate Model
lr_eval = evaluate_model(lr, X_test, y_test)

# Print result
print('Accuracy:', lr_eval['acc'])
print('Precision:', lr_eval['prec'])
print('Recall:', lr_eval['rec'])
print('F1 Score:', lr_eval['f1'])
print('Area Under Curve:', lr_eval['auc'])
print('Confusion Matrix:\n', lr_eval['cm'])
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Building KNN model 
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
# Evaluate Model
knn_eval = evaluate_model(knn, X_test, y_test)

# Print result
print('Accuracy:', knn_eval['acc'])
print('Precision:', knn_eval['prec'])
print('Recall:', knn_eval['rec'])
print('F1 Score:', knn_eval['f1'])
print('Area Under Curve:', knn_eval['auc'])
print('Confusion Matrix:\n', knn_eval['cm'])
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```
